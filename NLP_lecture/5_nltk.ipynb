### 데이터 정제

```python
import re
```

```python
emma_cleaned = re.sub('[^a-zA-Z \.\,]+', ' ', emma_raw)
emma_cleaned
```

### 토큰화

1. 문장 토큰화하기
2. 단어 토큰화하기

```python
from nltk import sent_tokenize
```

```python
sentences = sent_tokenize(emma_cleaned)

print('분리된 문장 개수 : ', len(sentences))
print(type(sentences))
print(sentences)
```

```python
sentences[100]
```

### 단어 토큰화하기

```python
import pandas as pd

docu_df = pd.DataFrame(data=sentences, columns=['sentence'])
print(docu_df.head())
```

```python
from nltk.tokenize import word_tokenize
```

```python
docu_df['word_tokenized'] = docu_df['sentence'].apply(lambda x : ','.join(word_tokenize(x)))
docu_df['word_tokenized'].head()
```

### 데이터 클렌징 2차

```python
docu_df['word_tokenized'] = docu_df['word_tokenized'].str.replace(',,,', ',')
docu_df['word_tokenized'].head()
```

```python
docu_df['word_tokenized'] = docu_df['word_tokenized'].str.lower()
docu_df['word_tokenized'].head()
```

### 불용어 제거

```python
# 데이터 확인하기

# Emma 수 확인
docu_df['word_tokenized'].str.count('emma').sum()
```

```python
word_tokenized_df = docu_df['word_tokenized'].str.split(',', expand=True)
word_tokenized_df
```

```python
tmp_df = docu_df['word_tokenized'].str.count(',')
docu_df['word_tokenized_cnt'] = tmp_df + 1
docu_df['word_tokenized_cnt']
```

```python
docu_df['word_tokenized_cnt'].min()
```

### 말뭉치 

```python
tmp_corpus = []
for col in word_tokenized_df.columns.tolist()[:5]:
    words = word_tokenized_df[col].tolist()
    print(f"col == {col} || word == {words}")
    tmp_corpus.extend(words)
```

```python
tmp_corpus
```

```python
tmp_corpus = []
for col in word_tokenized_df.columns.tolist():
    words = word_tokenized_df[col].tolist()
    tmp_corpus.extend(words)
```

```python
tmp_corpus_all = tmp_corpus.copy()
tmp_corpus_vocab = set(tmp_corpus)
len(tmp_corpus_all) , len(tmp_corpus_vocab)
```

### 불용어 적용하기

```python
stopwords = nltk.corpus.stopwords.words('english')
type(stopwords) , stopwords
```

```python
stopwords_set = set(stopwords)
global stopwords_set

def replace_stopword(word):
    words = word.split(',')
    words_set = set(words)
    rs_words = words_set.difference(stopwords_set)
    rs_words = ','.join(list(rs_words))
    return rs_words
   

docu_df['word_tokenized_fillteres'] = docu_df['word_tokenized'].apply(lambda x : replace_stopword(x))
```

```python
docu_df['word_tokenized_fillteres']
```

```python
word_tokenize_filtered_df = docu_df['word_tokenized_fillteres'].str.split(',', expand=True)

tmp_corpus_filtered = []
for col in word_tokenize_filtered_df.columns.tolist():
    words = word_tokenize_filtered_df[col].tolist()
    tmp_corpus_filtered.extend(words)

tmp_corpus_filtered_all = tmp_corpus_filtered.copy()
tmp_corpus_filtered_vocab = set(tmp_corpus_filtered)
len(tmp_corpus_filtered_all) , len(tmp_corpus_filtered_vocab)
```

```python
len(tmp_corpus_filtered_vocab) , len(tmp_corpus_vocab)
```

```python
counts_filtered = collections.Counter(tmp_corpus_filtered_all)
print(counts_filtered)
print(counts_filtered.most_common(15))
```

```python
print(counts.most_common(15))
```

### 어간 추출 / 표제어 추출

```python
from nltk.stem import PorterStemmer , LancasterStemmer
```

```python
st1 = PorterStemmer()
st2 = LancasterStemmer()

words = ['fly', 'flies', 'flying', 'flew', 'flown']

print("PorterStemmer : " , [st1.stem(w) for w in words] )
print("LancasterStemmer : " , [st2.stem(w) for w in words] )
```

표제어 추출

```python
from nltk.stem import WordNetLemmatizer
```

```python
lm = WordNetLemmatizer()
print([lm.lemmatize(w) for w in words])
```

```python
lm = WordNetLemmatizer()
print([lm.lemmatize(w, pos = 'v') for w in words])
```
